```python
## 원본 데이터 가져오기 

import pandas as pd 
trade_data = pd.read_csv('trade_data.csv')
main_data = pd.read_csv('main_data.csv')
ch_data = pd.read_csv('ch_data.csv')
sc_data = pd.read_csv('sc_data.csv')
```

```python
dp=pd.concat([trade_data,main_data,ch_data,sc_data])
```

```python
dp
```

```python
## store 테이블 생성 

store=pd.unique(dp['store'])
store=pd.DataFrame(store,columns=['store_name'],index=[i+1 for i in range(len(store))]).reset_index()
store.rename(columns={'index':'store_id'},inplace=True)
# store.to_csv('store.csv')
store
```

```python
#customre 테이블 생성 

custid=[]
for i in  dp['custid'].tolist():
    custid.append(i+1)
    
customer=pd.DataFrame({'custid': custid,'gender':dp['gender'].tolist()})
customer=customer.drop_duplicates(['custid','gender']).sort_values(['custid'])
# customer.to_csv('customer.csv' ,encoding='utf-8', index=False)
customer
```

```python
len(dp[['part']].drop_duplicates(['part']))
```

```python
## part 테이블 생성 

part=pd.unique(dp['part'])
part=pd.DataFrame(part,columns=['part_name'],index=[i+1 for i in range(len(part))]).reset_index()
part.rename(columns={'index':'part_id'},inplace=True)
part
# part.to_csv('part.csv',encoding='utf-8', index=False)
```

```python
#pc 테이블 생성 

pc=dp[['pc','part']].drop_duplicates(['pc']).reset_index(drop=True).reset_index()
pc=pc.rename(columns={'index':'pc_id' ,'part':'part_name','pc':'pc_name'})
pc=pd.merge(pc,part,how='left',on='part_name')
pc=pc.drop(['part_name'],axis=1)
pc['pc_id']=pc['pc_id']+1 
# pc.to_csv('pc.csv',encoding='utf-8', index=False)
pc 
```

```python
### corner 테이블 생성 

corner=dp[['corner','pc']].drop_duplicates(['corner']).rename(columns={'corner':
    'corner_name','pc':'pc_name'}).reset_index(drop=True).reset_index().rename(columns={'index':'corner_id'})
corner=pd.merge(corner,pc,how='left',on='pc_name').drop(['pc_name','part_id'],axis=1)
corner['corner_id']=corner['corner_id']+1
# corner.to_csv('corner.csv',encoding='utf-8', index=False)
corner
```

```python
## brand 테이블 생성 

brand=dp[['product','brand','corner']].drop_duplicates(['product','brand']).reset_index(drop=True).reset_index().rename(columns={'index'
    :'brand_id','product':'brandcode','corner':'corner_name','brand':'brand_name'})
brand=pd.merge(brand,corner,how='left',on='corner_name').drop(['corner_name','pc_id'],axis=1)
brand['brand_id']=brand['brand_id']+1 
# brand.to_csv('brand.csv',encoding='utf-8', index=False)
brand
```

```python
### transcation 테이블 생성 

store['store_id']=store['store_id']-1 
brand['brand_id']=brand['brand_id']-1

transaction=dp[['date_time','amount','discount','installment','product','brand','custid','store']].reset_index(drop=True).reset_index().rename(columns={'index'
:'transaction_id','product':'brandcode','brand':'brand_name','store':'store_name'})

transaction=pd.merge(transaction,brand,how='left',on=['brandcode','brand_name'])
transaction=pd.merge(transaction,store,how='left',on='store_name')


transaction=transaction.drop(['brandcode','brand_name','corner_id','store_name'],axis=1)

transaction_=transaction.drop_duplicates(['date_time','amount','discount','installment',
             'custid','brand_id','store_id'])
transaction['transaction_id']=transaction['transaction_id']+1
transaction['custid']=transaction['custid']+1

store['store_id']=store['store_id']+1 
brand['brand_id']=brand['brand_id']+1

transaction
```

```python
## transaction id 1부터 시작하게 맞추기 
transaction['brand_id']=transaction['brand_id']+1
transaction['store_id']=transaction['store_id']+1
transaction
```

```python
### erd 다이어그램 sql 전체 구문 
sql="""
CREATE TABLE IF NOT EXISTS `mydb`.`store` (
  `store_id` INT NOT NULL AUTO_INCREMENT,
  `store_name` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`store_id`))
ENGINE = InnoDB;

CREATE TABLE IF NOT EXISTS `mydb`.`customer` (
  `customer_id` INT NOT NULL AUTO_INCREMENT,
  `gender` INT NOT NULL,
  PRIMARY KEY (`customer_id`))
ENGINE = InnoDB;

CREATE TABLE IF NOT EXISTS `mydb`.`part` (
  `part_id` INT NOT NULL AUTO_INCREMENT,
  `part_name` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`part_id`))
ENGINE = InnoDB;



CREATE TABLE IF NOT EXISTS `mydb`.`pc` (
  `pc_id` INT NOT NULL AUTO_INCREMENT,
  `pc_name` VARCHAR(45) NOT NULL,
  `part_id` INT NOT NULL,
  PRIMARY KEY (`pc_id`),
  INDEX `fk_pc_part2_idx` (`part_id` ASC) VISIBLE,
  CONSTRAINT `fk_pc_part2`
    FOREIGN KEY (`part_id`)
    REFERENCES `mydb`.`part` (`part_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;



CREATE TABLE IF NOT EXISTS `mydb`.`corner` (
  `corner_id` INT NOT NULL AUTO_INCREMENT,
  `corner_name` VARCHAR(45) NOT NULL,
  `pc_id` INT NOT NULL,
  PRIMARY KEY (`corner_id`),
  INDEX `fk_corner_pc1_idx` (`pc_id` ASC) VISIBLE,
  CONSTRAINT `fk_corner_pc1`
    FOREIGN KEY (`pc_id`)
    REFERENCES `mydb`.`pc` (`pc_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;



CREATE TABLE IF NOT EXISTS `mydb`.`brand` (
  `brand_id` INT NOT NULL AUTO_INCREMENT,
  `brandcode` VARCHAR(45) NOT NULL,
  `brand_name` VARCHAR(45) NOT NULL,
  `corner_id` INT NOT NULL,
  PRIMARY KEY (`brand_id`),
  INDEX `fk_brand_corner1_idx` (`corner_id` ASC) VISIBLE,
  CONSTRAINT `fk_brand_corner1`
    FOREIGN KEY (`corner_id`)
    REFERENCES `mydb`.`corner` (`corner_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;



CREATE TABLE IF NOT EXISTS `mydb`.`transaction` (
  `transaction_id` INT NOT NULL AUTO_INCREMENT,
  `date_time` DATETIME NOT NULL,
  `amount` INT NOT NULL,
  `installment` INT NOT NULL,
  `discount` INT NOT NULL,
  `store_id` INT NOT NULL,
  `customer_id` INT NOT NULL,
  `brand_id` INT NOT NULL,
  PRIMARY KEY (`transaction_id`),
  INDEX `fk_transaction_store1_idx` (`store_id` ASC) VISIBLE,
  INDEX `fk_transaction_customer1_idx` (`customer_id` ASC) VISIBLE,
  INDEX `fk_transaction_brand1_idx` (`brand_id` ASC) VISIBLE,
  CONSTRAINT `fk_transaction_store1`
    FOREIGN KEY (`store_id`)
    REFERENCES `mydb`.`store` (`store_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_transaction_customer1`
    FOREIGN KEY (`customer_id`)
    REFERENCES `mydb`.`customer` (`customer_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_transaction_brand1`
    FOREIGN KEY (`brand_id`)
    REFERENCES `mydb`.`brand` (`brand_id`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
"""
```
